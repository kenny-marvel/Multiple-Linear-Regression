```python
import numpy as np
import pandas as pd
import scipy
import openpyxl
import statsmodels.api as sm
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
```

```python
raw_data=pd.read_csv(r'C:\Users\dell\Documents\1.03.+Dummies .csv')
raw_data
```

```python
raw_data.describe()
```

```python
data=raw_data.copy()
data['Attendance']=data['Attendance'].map({'Yes':1,'No':0})
data
```

```python
data.describe()
```

```python
y=data['GPA']
x1=data[['SAT','Attendance']]
```

```python
x=sm.add_constant(x1)
result=sm.OLS(y,x).fit()
result.summary()
```

```python
#plot the regression on scatter plot
plt.scatter(data['SAT'],y)
yhat_no=0.6439+0.0014*data['SAT']
yhat_yes=0.8665+0.0014*data['SAT']
#plot the regression line
fig=plt.plot(data['SAT'],yhat_no,lw=2,c='red')
fig=plt.plot(data['SAT'],yhat_yes,lw=2,c='green')
#name your axes
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
## plot the regression on scatter plot and color the data point
#the color depends on the attendance
plt.scatter(data['SAT'],data['GPA'],c=data['Attendance'],cmap='coolwarm')
#Define the two regression equation(one with a dummy=1,other dummmy=0)
yhat_no=0.6439+0.0014*data['SAT']
yhat_yes=0.8665+0.0014*data['SAT']
#plot the two regression line
fig=plt.plot(data['SAT'],yhat_no,lw=2,c='red')
fig=plt.plot(data['SAT'],yhat_yes,lw=2,c='blue')
#plot the axes
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
#add the original regression line
#color depend on attendance
#create one scatter plot that contains all observation
#use the series 'attendance as color' and choose  a color map of your choice
plt.scatter(data['SAT'],data['GPA'],c=data['Attendance'],cmap='coolwarm')
#Define the two regression equation(one with a dummy=1,other dummmy=0)
yhat_no=0.6439+0.0014*data['SAT']
yhat_yes=0.8665+0.0014*data['SAT']
#Original regression line
yhat=0.0017*data['SAT']+0.275
#plot the two regression line
fig=plt.plot(data['SAT'],yhat_no,lw=2,c='red',label='Regression line 1')
fig=plt.plot(data['SAT'],yhat_yes,lw=2,c='blue',label='Regression line 2')
#PLOT THE ORIGINAL REGRESSION LINE
fig=plt.plot(data['SAT'],yhat,lw=3,c='blue',label='Regression line ')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.legend()
plt.show()
```

```python
#MAKING PREDICTION WITH THE LINEAR REGRESSION CREATED
#created a new data frame,identical in organisation to x
#the constant is always 1,while each of the line correspond to an observation(student)
new_data=pd.DataFrame({'const':1,'SAT':[1700,1670],'Attendance':[0,1]})
new_data
#by default, when you create a df(not load,but create),the column are sorted alphabetically
#so,if we dont reorder them,they would be 'Attendance','const','SAT'
#if you feed them in the wrong order,wrong result will be given
```

```python
#renaming the indices for the purpose of this example
#that's not really a good practice => it wont overwrite the variable
#using NUMPY, SKlearn etc method on a df with renamed indices, they will simply be lost
#and returned to 0,1,2,3 etc
new_data.rename(index={0:'Bob',1:'Alice'})
#use the predict method on the regression with the new data as a single argument
predictions=result.predict(new_data)
predictions
```

```python
#if you want, we can create a dataframe including everything
predictionsdf=pd.DataFrame({'predictions':predictions})
#join the two dataframe
joined=new_data.join(predictionsdf)
#rename the indices as before(not a good practice in general)
joined.rename(index={0:'Bob',1:'Alice'})
```
